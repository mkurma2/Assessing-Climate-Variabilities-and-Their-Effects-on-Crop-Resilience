```python
import pandas as pd
import numpy as np
import seaborn as sns
%matplotlib inline
import matplotlib.pyplot as plt
from matplotlib import rcParams
```

```python
disaster_data = pd.read_csv(r"time-series-disasters-1980-2022.csv")
```

```python
disaster_data
```

```python
ax = sns.barplot(x="Year", y="All Disasters Count", data=disaster_data)
rcParams['figure.figsize'] = 28,14
plt.ylabel('Disaster Count')
plt.show()
```

We can see a enormous rise in the natural disaster count as the years pass by. This has a direct correlation with the increasing temperature values. We also know that these disasters have an adverse impact on crop production and the increasing number is a cause of concern.
